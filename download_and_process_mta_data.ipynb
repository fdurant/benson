```python
import sys
import time
from datetime import datetime, timedelta, date
import os
import os.path
import urllib2

def getDateList(maxDate):
    dates = []
    # MTA started collecting on Wednesday May 5th 2010
    dates.append(datetime(2010, 5, 5, 0, 0, 0))
    # Other exception
    dates.append(datetime(2011, 12 ,17, 0, 0, 0))
    # After that, they collect each and every Saturday until now, starting on Sat May 8th 2010
    firstSaturday = datetime(2010, 5, 8, 0, 0, 0)
    # Uncomment next line for debugging (fewer files & downloads)
    # firstSaturday = datetime(2014, 12, 27, 0, 0, 0)

    currentDate = firstSaturday
    nrWeeksFromStart = 0
    while True:
        currentDate = firstSaturday + timedelta(weeks = nrWeeksFromStart)
        if currentDate > today:
            break
        dates.append(currentDate)
        nrWeeksFromStart += 1
    return dates

def downLoadFilesUnlessAlreadyDone(dates):
    baseMtaUrl = "http://web.mta.info/developers/data/nyct/turnstile/"
    turnstileFileNameBase = "turnstile_"
    turnstileFileNameExt = ".txt"
    
    for d in dates:
        turnstileFileName = "%s%s%s" % (turnstileFileNameBase, d.strftime("%y%m%d"), turnstileFileNameExt)
        if os.path.isfile(turnstileFileName) and os.stat(turnstileFileName).st_size > 0:
            print >> sys.stderr, "Already have %s, so ignoring!" % turnstileFileName
        else:
            url = "%s%s" % (baseMtaUrl, turnstileFileName)
            print >> sys.stderr, "Attempting download of %s" % url
            try:
                onlineFile = urllib2.urlopen(url)
                localFile = open(turnstileFileName,"w")
                localFile.write(onlineFile.read())
                localFile.close()
                onlineFile.close()
            except Exception:
                print >> sys.stderr, str(Exception)
    print >> sys.stderr, "Download ready"
                
# Make list of all dates to download
today = datetime.today()
downloadDates = getDateList(today)
#print >> sys.stderr, "\n".join([str(date) for date in downloadDates])
downLoadFilesUnlessAlreadyDone(downloadDates)
```

```python
from datetime import datetime

def groupValuesPerDateTime(dict):
    resultDict = {}
    for k,v in dict.items():
        for elem in v:
            mm, dd, yyyy = elem[2].split('/')
            h, m, s = elem[3].split(':')
            dt = datetime(int(yyyy),int(mm),int(dd),int(h),int(m),int(s))
#            print >> sys.stderr, elem
            try:
                cumNrEntries = int(elem[5])
            except Exception:
                print >> sys.stderr, "elem is %s" % str(elem)
            if not k in resultDict:
                resultDict[k] = []
            resultDict[k].append([dt,cumNrEntries])
    return resultDict
turnstileDateTimeDict = groupValuesPerDateTime(turnstileDict)

print >> sys.stderr, 'turnstileDateTimeDict has %i elements' % len(turnstileDateTimeDict)
```

### Challenge 3

```python
def groupValuesPerDay(dict):
    resultDict = {}
    for k,v in dict.items():
        perDayList = []
        vSortedPerDatetime = sorted((elem for elem in v), key=lambda elem: elem[0])
        #print >> sys.stderr, vSortedPerDatetime
        smallestDatetime = vSortedPerDatetime[0][0]
        smallestDate = datetime(smallestDatetime.year,smallestDatetime.month,smallestDatetime.day)
        previousDate = smallestDate
        previousCumNrEntries = vSortedPerDatetime[0][1]
        absNrEntriesThisDay = 0
        for elem in vSortedPerDatetime:
            dt = elem[0]
            dayKey = datetime(dt.year,dt.month,dt.day)
            cumNrEntries = elem[1]
            if dayKey > previousDate:
                # This is a new day, so store previous result and reinitialize
                perDayList.append([dayKey, absNrEntriesThisDay])
                absNrEntriesThisDay = 0
                previousDate = dayKey
            if dayKey <= smallestDate:
                # We have no previous cumulative figure to compare against, so discard this day from the output
                pass
            else:
                # To get absolute figures, take difference between cumulative figures
                absNrEntriesPerMeasurement = cumNrEntries - previousCumNrEntries
                try:
                    #assert(absNrEntriesPerMeasurement >= 0), 'absNrEntriesPerMeasurement %d is negative!' % absNrEntriesPerMeasurement
                    assert(absNrEntriesPerMeasurement < 100000), 'absNrEntriesPerMeasurement %d is ridiculously high!' % absNrEntriesPerMeasurement
                    absNrEntriesThisDay += absNrEntriesPerMeasurement
                except AssertionError:
                    print >> sys.stderr, "k = %s; absNrEntriesPerMeasurement = %s (%d - %d) found in %s" % (k,absNrEntriesPerMeasurement,cumNrEntries,previousCumNrEntries,elem)
                    previousCumNrEntries = cumNrEntries # Make sure we can at least continue just after this one
                    next
            previousCumNrEntries = cumNrEntries
        resultDict[k] = perDayList  
    return resultDict
turnstilePerDayDict = groupValuesPerDay(turnstileDateTimeDict)

print >> sys.stderr, 'turnstilePerDayDict has %i elements' % len(turnstilePerDayDict)
print >> sys.stderr, 'turnstilePerDayDict[%s] is %s' % (demoKey,turnstilePerDayDict[demoKey])
#for k in turnstilePerDayDict.keys()[0:1]:
#    print >> sys.stderr, 'turnstilePerDayDict[%s] is %s' % (k,turnstilePerDayDict[k])
```

```python
def groupValuesPerControlAreaUnitStation(dict):
    resultDict = {}
    tmpDict = {}
    for k,v in dict.items():
        # We don't need the 'SCP' column, i.e. k[2] anymore
        newKey = (k[0],k[1],k[3])
        if not newKey in tmpDict:
            tmpDict[newKey] = {} # Work with a nested dict
        for elem in v:
            if not elem[0] in tmpDict[newKey]:
                tmpDict[newKey][elem[0]] = 0
            tmpDict[newKey][elem[0]] += elem[1]
    # Convert the nested dict into a list sorted by datetime
    for newKey,newValue in tmpDict.items():
        sortedList = []
        for dt in sorted(newValue.keys()):
            sortedList.append([dt,newValue[dt]])
        resultDict[newKey] = sortedList
    return resultDict
turnstilesPerControlAreaUnitStationDict = groupValuesPerControlAreaUnitStation(turnstilePerDayDict)

print >> sys.stderr, 'turnstilesPerControlAreaUnitStationDict has %i elements' % len(turnstilesPerControlAreaUnitStationDict)
print >> sys.stderr, 'turnstilesPerControlAreaUnitStationDict[%s] is %s' % ((demoKey[0],demoKey[1],demoKey[3]),turnstilesPerControlAreaUnitStationDict[(demoKey[0],demoKey[1],demoKey[3])])
#for k in turnstilesPerControlAreaUnitStationDict.keys()[0:1]:
#    print >> sys.stderr, 'turnstilesPerControlAreaUnitStationDict[%s] is %s' % (k,turnstilesPerControlAreaUnitStationDict[k])
```

```python
def groupValuesPerStation(dict):
    resultDict = {}
    tmpDict = {}
    for k,v in dict.items():
        # We don't need the 'C/A' and 'UNIT' columns anymore
        newKey = k[2] # No need for a tuple anymore
        if not newKey in tmpDict:
            tmpDict[newKey] = {} # Work with a nested dict
        for elem in v:
            if not elem[0] in tmpDict[newKey]:
                tmpDict[newKey][elem[0]] = 0
            tmpDict[newKey][elem[0]] += elem[1]
    # Convert the nested dict into a list sorted by datetime
    for newKey,newValue in tmpDict.items():
        sortedList = []
        for dt in sorted(newValue.keys()):
            # Safeguard against rubbish
            if newValue[dt] > 0 and newValue[dt] < 300000:
                sortedList.append([dt,newValue[dt]])
        resultDict[newKey] = sortedList
    return resultDict
turnstilesPerStationDict = groupValuesPerStation(turnstilesPerControlAreaUnitStationDict)

print >> sys.stderr, 'turnstilesPerStationDict has %i elements' % len(turnstilesPerStationDict)
print >> sys.stderr, 'turnstilesPerStationDict[%s] is %s' % (demoKey[3],turnstilesPerStationDict[demoKey[3]])
#for k in turnstilesPerStationDict.keys()[0:1]:
#    print >> sys.stderr, 'turnstilesPerStationDict[%s] is %s' % (k,turnstilesPerStationDict[k])
```

```python
def mapWeekday2MyIndex(dateTimeWeekday):
    if dateTimeWeekday == 6:
        return 0
    else:
        if dateTimeWeekday < 6:
            return dateTimeWeekday + 1
        else:
            assert(False), "dateTimeWeekday value must lie between 0 and 6"

def groupValuesPerStationPerWeek(dict):
    resultDict = {}
    for k,v in dict.items():
        # Assumption: v is a list with list elements, sorted by datetime
        # the datetime of the first element is a Sunday
        listOfWeeks = []
        weekdayList = [0,0,0,0,0,0,0]
        for elem in v:
            #print >> sys.stderr, elem
            dt = elem[0]
            nrEntries = elem[1]
            if dt.weekday() == 6: # Sunday = first day of week
                # Prepare a data structure to store the data for a new week,
                # starting on Sunday and ending six days later on a Saturday
                weekEndingOnSaturdayKey = dt + timedelta(days = 6)
            weekdayList[mapWeekday2MyIndex(dt.weekday())] = nrEntries
            if dt.weekday() == 5: # Saturday = last day of week
                listOfWeeks.append([dt,weekdayList])
                weekdayList = [0,0,0,0,0,0,0] # re-initialize
        resultDict[k] = listOfWeeks
    return resultDict
turnstilesPerStationPerWeekDict = groupValuesPerStationPerWeek(turnstilesPerStationDict)

print >> sys.stderr, 'turnstilesPerStationPerWeekDict has %i elements' % len(turnstilesPerStationPerWeekDict)
#print >> sys.stderr, "\n".join(turnstilesPerStationPerWeekDict.keys())
    
randomStationKey = turnstilesPerStationPerWeekDict.keys()[random.randint(0,len(turnstilesPerStationPerWeekDict))]
randomStationKey = demoKey[3]
print >> sys.stderr, 'turnstilesPerStationPerWeekDict[%s] is %s' % (randomStationKey,turnstilesPerStationPerWeekDict[randomStationKey])
#for k in turnstilesPerStationPerWeekDict.keys()[0:1]:
#    print >> sys.stderr, 'turnstilesPerStationPerWeekDict[%s] is %s' % (k,turnstilesPerStationPerWeekDict[k])

plt.figure(figsize=(15,10))
plt.title('Weekly entries in MTA station: %s' % str(randomStationKey))
plt.xlabel('Days of week (Sunday through Saturday)')
plt.ylabel('Number of entries')
for elem in turnstilesPerStationPerWeekDict[randomStationKey][1:]:
    weekdays = [0,1,2,3,4,5,6]
    counts = elem[1]
    plt.plot(weekdays,counts, label='week ending on %s' % elem[0].strftime("%a %m/%d/%Y"))
plt.legend()
```
